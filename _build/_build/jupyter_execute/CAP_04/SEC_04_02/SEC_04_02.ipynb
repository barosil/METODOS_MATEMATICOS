## DIstribuição Binomial

```{prf:definition}
Se uma variável aleatória $X$ contém o número de tentativas que resultam em sucesso para um evento individual cuja probabilidade de sucesso é $p$ e fracasso $1-p$ escrevemos que $X\sym B(n, p)$ se a distribuição de $X$ é uma distribuição binomial. A probabilidade de obter $k$ sucessos é dada por 
$$
f(k;n, p) = \binom(n, k) \;p^k\; (1-p)^{n-k}
$$
```
